```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('http://files.grouplens.org/datasets/movielens/ml-100k/u.data')
df
```

```python
df = pd.read_csv(
    'http://files.grouplens.org/datasets/movielens/ml-100k/u.data', 
    names=["user_id", 
           "item_id", 
           "review", 
           "timestamp"], 
    sep="\t"
)
df
```

```python
df1 = pd.pivot_table(df.drop("timestamp", axis=1), index="user_id", columns="item_id", values="review")
df1
```

```python
df2 = df1.fillna(0)
```

```python
#array[0][0] ⇒ 行番号・列番号
#df.iloc[0, 0] ⇒ 行位置・列位置
#df.loc[行ラベル, 列ラベル] ⇒ ラベル名でのアクセス
cosine_similarity([df2.iloc[0]], [df2.iloc[1]])
#df2の0行目とdf2の1行目をcosineの類似度で
```

```python
#空のリストを作る
cos_list = []


for user_id in df2.index:
    #index名:2/column名:2以降と指定したuser_idの評価2以上のもののcosine類似度を算出して、cos_listに追加(append)
    cos_list.append(cosine_similarity([df2.loc[2, 2:]], [df2.loc[user_id, 2:]]))
cos_list
```

```python
#cos_listの数を数えて(len)、順番付けして(range)、ソートする
cos_list1 = sorted(range(len(cos_list)), 
                   #cos_listを基準に、range(len(cos_list))を並び替えたい
                   key=lambda i: cos_list[i], reverse=True)[:51]
cos_list1
```

```python
#優先度順に並んだcos_listのインデックス番号の順に、df2のインデックス番号をcolumnと
df_similality = df2.iloc[cos_list1[1:]]
df_similality
```

```python
df_similality[df_similality[1] != 0.0][1].mean()
```
